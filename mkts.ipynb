# Get the data

Transform into 10 fold validation, each 6 days long.

```python
import statistics as st
def make(df,med):
    x = []
    df["ATR"] = df["Open"]
    for i,row in df.iterrows():
        x.append(st.mean([row["Open"], row["High"],row["Low"], row["Close"]] ))
        df.at[i,"ATR"] = 0 if len(x) <2 else st.stdev(x)
        if len(x) >med:
            x.pop(0) 
    return df
```

```python
import yfinance as yf
import pandas as pd
from datetime import date
from datetime import timedelta
# Get the data
datas = []
today = date.today()
for i in range(1,10):
    data = yf.download(tickers="EURUSD=X", interval="5m", start = (today - (timedelta(days=i*6))), end= (today - ((timedelta(days=(i-1)*6)))))
    datas.append(make(data,15))
```
